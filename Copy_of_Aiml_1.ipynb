```python
people = ['ammu','baby','kutty','sree','srilatha','vaasu','sneha','srinidi','siri','tharuni','teju','ashu']
age = [18,15,19,21,30,12,14,25,23,32,17,27]
weight = [30,43,45,62,23,50,35,27,37,35,75,49]
height = [122,233,188,199,144,177,211,233,901,222,333,444]
```

```python
import matplotlib.pyplot as plt #creating static,animated and interactive visualization.
```

```python
plt.scatter(weight,height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("peoples weight in kgs",fontsize=20,fontstyle='italic')

#main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6,color='red',edgecolor='k',alpha=0.6)
#edgecolor='k': sdets the edge color of the bars to black
#alpha=0.6: sets the transparency of the bars,with 1 being fully opaque and 0 fully transpar

plt.xlabel("peoples",fontsize=15)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))

plt.hist(weight,color='purple',edgecolor='k',alpha=0.75,bins=5)

plt.title("histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.show()
```
